```python
import sys

sys.path.append("..")

from src.agents.columns_analyzer import MySQLTableProfiler
from datetime import datetime

import asyncmy
import greenlet
import pandas as pd
from sqlalchemy.engine import URL
from sqlalchemy import create_engine, text

url = URL.create(
    drivername="mysql+asyncmy",
    username="root",
    password="your_db_password",
    host="your_db_host",
    port=3306,
    database="bdl",
)

engine = create_engine(url)

profiler = MySQLTableProfiler(
        dsn=url,
        schema=None
    )
```

```python
desc_table = pd.read_csv('../data/full_desc_table_v2.csv')
desc_table.head()
```

```python
tables = desc_table['TABLE'].tolist()
print(tables)
```

```python
desc_cols = []
for t in tables:
    if t == "BDL.D_LOAN_TYP_T":
        desc_cols.append("")
        continue
    report = await profiler.profile_table(t.replace("BDL.", "").lower(), top_k=50)
    desc_cols.append(await profiler.format_profile(report["columns"], max_values=None))
```
